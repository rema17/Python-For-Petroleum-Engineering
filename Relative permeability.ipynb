**import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

**Save data into datafram using pandas** 




```python
data = pd.DataFrame({'Sw': [0.3, 0.4, 0.5, 0.6, 0.7],
                      'krw': [0.0000,0.0004,0.0039,0.0157,0.0466],
                      'kro' : [1.0000,0.7195,0.4858,0.2985,0.1574]})
print(data)
                      
```

**plot relative permeability curve**


this plot used matplotlib

```python
fig, ax1 = plt.subplots()

ax1.plot(data.Sw, data.kro, 'g-')
ax1.set_xlabel('Sw')
ax1.set_ylabel('kro', color='green')

ax2 = ax1.twinx()
ax2.plot(data.Sw, data.krw)
ax2.set_ylabel('krw', color='blue')
```

**plot using Plotly**

```python
import plotly.graph_objects as go

# Create the figure
fig = go.Figure()

# Plot 'kro' on the left y-axis
fig.add_trace(go.Scatter(x=data.Sw, y=data.kro, mode='lines', name='kro', line=dict(color='green')))

# Plot 'krw' on the right y-axis
fig.add_trace(go.Scatter(x=data.Sw, y=data.krw, mode='lines', name='krw', line=dict(color='blue'), yaxis="y2"))

# Add axis labels
fig.update_layout(
    xaxis=dict(title="Sw"),
    yaxis=dict(title="kro", titlefont=dict(color="green")),
    yaxis2=dict(title="krw", titlefont=dict(color="blue"), overlaying="y", side="right")
)

# Show the figure
fig.show()
```

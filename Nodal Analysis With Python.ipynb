# Nodal Analysis With Python



```python

import pandas as pd 
import numpy as np
import matplotlib as plt 
import seaborn as sns
%matplotlib inline

```

```python
#data sourse from petroWiki


#Create Datafream 

IPR = pd.DataFrame({'Pwf':[4000,3500,3000,2500,2000,1500,1000,500,14.7],
                    'Q': [0,1999,3094,3902,4512,4963,5275,5458,5519]})

IPR
```

```python
#Now TPR Data

q = np.arange(1000, 6500, 500)

data = {
    'q': q,
    'Pwf(1.90" tbg)': [1334, 1400, 1487, 1592, 1712, 1843, 1984, 2132, 2287, 2446, 2609],
    'Pwf(2.375"tbg)': [1298, 1320, 1351, 1390, 1435, 1487, 1545, 1609, 1677, 1749, 1824],
    'Pwf(2.875"tbg)': [1286, 1294, 1305, 1319, 1336, 1356, 1378, 1403, 1431, 1461, 1493]
}

TPR = pd.DataFrame(data)
```

```python
TPR
```

```python

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(IPR['Q'], IPR['Pwf'], label='IPR', linewidth=1.5)

ax.plot(TPR['q'], TPR['Pwf(1.90" tbg)'], label='TPR for 1.9" tubing', linewidth=1.5)
ax.plot(TPR['q'], TPR['Pwf(2.375"tbg)'], label='TPR for 2.375" tubing', linewidth=1.5)
ax.plot(TPR['q'], TPR['Pwf(2.875"tbg)'], label='TPR for 2.875" tubing', linewidth=1.5)

ax.set_xlabel('Flow Rate (Mscf/D)')
ax.set_ylabel('Flowing BHP (Psia)')
ax.set_title('Nodal Analysis for a Gas Well')

ax.grid()

ax.legend(loc='best')
plt.savefig('nodal_analysis_plot1.png', dpi=300, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(IPR['Q'], IPR['Pwf'], label='Inflow Performance Relationship (IPR)', linewidth=1.5, color='black')

ax.plot(TPR['q'], TPR['Pwf(1.90" tbg)'], label='Tubing Performance Relationship (TPR) for 1.9" tubing', linewidth=1.5, color='red')
ax.plot(TPR['q'], TPR['Pwf(2.375"tbg)'], label='TPR for 2.375" tubing', linewidth=1.5, color='green')
ax.plot(TPR['q'], TPR['Pwf(2.875"tbg)'], label='TPR for 2.875" tubing', linewidth=1.5, color='blue')

ax.set_xlabel('Flow Rate (Mscf/D)', fontsize=12)
ax.set_ylabel('Flowing Bottomhole Pressure (Psia)', fontsize=12)
ax.set_title('Nodal Analysis for a Gas Well', fontsize=14, fontweight='bold')

ax.grid(alpha=0.5)

ax.legend(loc='best', fontsize=10)

sns.despine()
plt.savefig('nodal_analysis_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

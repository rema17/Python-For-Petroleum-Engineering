**import libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**upload dataset**

```python
data = pd.read_csv("E:\\Teaching\\Online courses\\python for petroleum engineering\\Data\\L0509WellData.csv")
data
```

**create well log plot**

```python
fig, ax=plt.subplots(nrows=1, ncols=5, figsize=(16,12), sharey =True) # this line will create 1 row and 5 columns 
fig.suptitle(" well log plot", fontsize=30) # this line write title for the plot

#adjust for the log tracks 
#Track one 
TK1 = data['GR']
TK2 = data['DEPTH']
ax1=ax[0].twiny() # ax[0] plot the first column, .twiny()" is a method to share the y-axis.
ax1.invert_yaxis()

ax1.plot(TK1, TK2, color= 'Green', linestyle='-') #this will plot the GR on the x-axis and depth for y-axis
ax1.set_xlabel('GR',color= 'Green') #this will assign x label and color
ax1.spines['top'].set_position(('outward',1)) ## This indicates data area boundary and adjusting location of label

#Track Two
TK3 = data['DT']
TK4 = data['DEPTH']

ax2 = ax[1].twiny() 
ax2.invert_yaxis()
ax2.plot(TK3, TK4, color= 'blue', linestyle='-')
ax2.set_xlabel('DT',color= 'blue')
ax2.tick_params(axis='x', color='blue') # This will adjust tick label, color, and axis on which to apply the tick label
ax2.spines['top'].set_position(('outward',1))

### Hint : DT log contain -999 value which is NaN value "stright line on the top 

#Track Three
TK5 = data['RHOB']
TK6 = data['DEPTH']

ax3 = ax[2].twiny() 
ax3.invert_yaxis()
ax3.plot(TK5, TK6, color= 'red', linestyle='-')
ax3.set_xlabel('RHOB',color= 'red')
ax3.tick_params(axis='x', color='red')
ax3.spines['top'].set_position(('outward',1))


#Track Four
TK7 = data['DRHO']
TK8 = data['DEPTH']

ax4 = ax[3].twiny() 
ax4.invert_yaxis()
ax4.plot(TK7, TK8, color= 'red', linestyle='-')
ax4.set_xlabel('DRHO',color= 'red')
ax4.tick_params(axis='x', color='red')
ax4.spines['top'].set_position(('outward',1))


#Track Five
TK9 = data['NPHI']
TK10 = data['DEPTH']

ax5 = ax[4].twiny() 
ax5.invert_yaxis()
ax5.plot(TK9, TK10, color= 'black', linestyle='-')
ax5.set_xlabel('NPHI',color= 'black')
ax5.tick_params(axis='x', color='black')
ax5.spines['top'].set_position(('outward',1))
```
